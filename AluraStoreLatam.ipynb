### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#tienda 1
tienda_total_de_ventas1 = tienda["Precio"].sum()
print(f"La tienda 1 recaudó un total de {tienda_total_de_ventas1}")
#tienda2
tienda_total_de_ventas2 = tienda2["Precio"].sum()
print(f"La tienda 2 recaudó un total de {tienda_total_de_ventas2}")
#tienda 3
tienda_total_de_ventas3 = tienda3["Precio"].sum()
print(f"La tienda 3 recaudó un total de {tienda_total_de_ventas3}")
#tienda 4
tienda_total_de_ventas4 = tienda4["Precio"].sum()
print(f"La tienda 4 recaudó un total de {tienda_total_de_ventas4}")

# Crear la gráfica
plt.bar(["Tienda 1","Tienda 2","Tienda 3","Tienda 4"], [tienda_total_de_ventas1,tienda_total_de_ventas2,tienda_total_de_ventas3,tienda_total_de_ventas4], color='green')
plt.ylabel("Total Recaudado")
plt.title("Total de Ventas - Tienda ")
plt.show()
```

```python
print(tienda_total_de_ventas3-tienda_total_de_ventas4)
```

```python
ventas = [tienda_total_de_ventas1, tienda_total_de_ventas2, tienda_total_de_ventas3, tienda_total_de_ventas4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(ventas, labels=tiendas, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)
plt.title("Porcentaje del Total Recaudado por Tienda")
plt.axis("equal")
plt.show()
```

# 2. Ventas por categoría

```python
#tienda1
Categorias_tienda1 = tienda["Categoría del Producto"].value_counts()
#tienda2
Categorias_tienda2 = tienda2["Categoría del Producto"].value_counts()
#tienda3
Categorias_tienda3 = tienda3["Categoría del Producto"].value_counts()
#tienda4
Categorias_tienda4 = tienda4["Categoría del Producto"].value_counts()

# Unir en un DataFrame
df_categorias = pd.DataFrame({
    "Tienda 1": Categorias_tienda1,
    "Tienda 2": Categorias_tienda2,
    "Tienda 3": Categorias_tienda3,
    "Tienda 4": Categorias_tienda4
}).fillna(0)

# Graficar
df_categorias.plot(kind='bar', figsize=(10,6))
plt.title("Categorías por Tienda")
plt.ylabel("Cantidad de productos")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
